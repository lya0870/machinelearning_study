```python
import pandas as pd

train=pd.read_csv(r'C:\Users\IBK\Desktop\PythonWorkspace\1주차_Titanic\train.csv')
test=pd.read_csv(r'C:\Users\IBK\Desktop\PythonWorkspace\1주차_Titanic\test.csv')
```

```python
#Age
#전체 나이 평균
train['Age']=train['Age'].fillna(train['Age'].mean())
#결측값을 채우는 방법은 다양하다. 
train.head()
```

```python
#결측값을 채우는 방법은 다양하다. 
#-결측값을 특정 값으로 채우기
#-결측값을 앞 방향 혹은 뒷 방향으로 채우기
#-결측값 채우는 횟수를 제한하기
#-결측값을 변수별 평균으로 대체하기
#-결측값을 다른 변수의 값으로 대체하
#이 관련한 내용은 '결측값 채우는 방법' 주피터에서 기록
```

```python
test['Age']=test['Age'].fillna(test['Age'].mean());
#마지막에 ;을 뒤에 붙이면 out에 결과표시가 생략된다고 함.
```

```python
train.loc[train['Age']<10,'Age']=0
train.loc[(train['Age'] >= 10) & (train['Age'] < 20), 'Age'] = 1
train.loc[(train['Age'] >= 20) & (train['Age'] < 30), 'Age'] = 2
train.loc[(train['Age'] >= 30) & (train['Age'] < 40), 'Age'] = 3
train.loc[(train['Age'] >= 40) & (train['Age'] < 50), 'Age'] = 4
train.loc[train['Age'] >= 50, 'Age'] = 5
```

```python
#Family Size
train['FamilySize']=train['SibSp'] + train['Parch']
train.head()
```

```python
# 테스트셋도 똑같이 적용
test['FamilySize'] = test['SibSp'] + test['Parch']
```

```python
# 더이상 필요 없는 열은 버리고, 필요한 필드만 남기기
new_train = train[['Survived', 'Sex', 'Age', 'FamilySize']]
new_test = test[['Sex', 'Age', 'FamilySize']] 
# test데이터는 애초에 Survived가 없음

train.head()
```

```python
#Fare
#운임도 빈칸(NaN)을 운임평균으로 채우기
train['Fare'] = train['Fare'].fillna(train['Fare'].mean())
test['Fare'] = test['Fare'].fillna(test['Fare'].mean())
```

```python
train.isnull().sum()
```

```python
train['Embarked'].value_counts()
```

```python
#Embarked
train['Embarked'] = train['Embarked'].fillna('S')
test['Embarked'] = test['Embarked'].fillna('S')
```

```python
# 필요한 필드만 남기기
new_train = train[['Survived', 'Sex', 'Age', 'FamilySize', 'Fare', 'Embarked']]
new_test = test[['Sex', 'Age', 'FamilySize', 'Fare', 'Embarked']] 
# test데이터는 애초에 Survived가 없음

new_train.head()
```

```python
#Feature to Numeric
train.loc[train['Sex'] == 'male', 'Sex'] = 0
train.loc[train['Sex'] == 'female', 'Sex'] = 1

test.loc[test['Sex'] == 'male', 'Sex'] = 0
test.loc[test['Sex'] == 'female', 'Sex'] = 1

train.head()
```

```python
train.loc[train['Embarked'] == 'S', 'Embarked'] = 0
train.loc[train['Embarked'] == 'C', 'Embarked'] = 1
train.loc[train['Embarked'] == 'Q', 'Embarked'] = 2

test.loc[test['Embarked'] == 'S', 'Embarked'] = 0
test.loc[test['Embarked'] == 'C', 'Embarked'] = 1
test.loc[test['Embarked'] == 'Q', 'Embarked'] = 2

train.head()
```

```python
new_train = train[['Survived', 'Sex', 'Age', 'FamilySize', 'Fare', 'Embarked']]
new_test = test[['Sex', 'Age', 'FamilySize', 'Fare', 'Embarked']] 

new_train.head()
```

```python
#Modeling
x_train = new_train[['Sex', 'Age', 'FamilySize', 'Fare', 'Embarked']]
y_train = new_train['Survived'] 
# 선택할 열이 하나면, []를 한번만 쓰기
x_train
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier()
tree.fit(x_train, y_train)

print('training set accuracy:', tree.score(x_train, y_train))
```

```python
#Prediction

x_test = new_test[['Sex', 'Age', 'FamilySize', 'Fare', 'Embarked']]
# 사실 이 작업을 안해도 new_test와 동일하기 때문에 생략가능

x_test
```

```python
prediction = tree.predict(x_test) 
# 즉 y_test가 생성됨
prediction
```

```python
#csv 파일로 만들기
submit = pd.DataFrame({'PassengerId': test['PassengerId'],'Survived': prediction})
submit.to_csv('submit.csv', index=False)
```
